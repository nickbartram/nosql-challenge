### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}
results = list(establishments.find(query))

# Use count_documents to display the number of documents in the result
count_hygiene = establishments.count_documents(query)
print(f"The number of establishments with a hygiene score of 20 is: {count_hygiene}.")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(len(hygiene_df))

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Set the number of degrees to search 
degree_search = 0.01

# Dynamically set the latitude and longitude for "Penang Favlours"
latitude = establishments.find_one({'BusinessName': 'Penang Flavours'},
                               {'geocode.latitude': 1, '_id':0})['geocode']['latitude']

longitude = establishments.find_one({'BusinessName': 'Penang Flavours'},
                               {'geocode.longitude': 1, '_id':0})['geocode']['longitude']

# Create the query for Rating Value and longitude/latitude
geo_query = {'RatingValue': 5,
             'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
             'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude  + degree_search}}
geo_sort = [('scores.Hygiene', 1)]

# Set limit to 5 for the Top 5 establishments
geo_limit = 5

# Save results to a list
geo_results = list(establishments.find(geo_query).sort(geo_sort).limit(geo_limit))

# Print results by looping throughthe list
for result in geo_results:
    pprint(result)

```

```python
# Convert result to Pandas DataFrame
geo_df = pd.DataFrame(geo_results)
geo_df.head(10)
```

Create a DataFrame with reduced columns that are more relevant to the queries. This is an optional choice that was not specified in the the instructions.

```python
# Use json_normalize to access longitude, latitude, and hygiene score directly (not nested)
reduced_geo_df = pd.json_normalize(geo_results)

# Reduce columns to relevant fields
reduced_geo_df = reduced_geo_df[['_id', 'BusinessName', 'RatingValue', 'scores.Hygiene', 'geocode.latitude', 'geocode.longitude']]

# Rename columns for clarity
reduced_geo_df = reduced_geo_df.rename(columns= {'BusinessName': 'Business Name',
                                                 'scores.Hygiene': 'Hygiene Score',
                                                 'RatingValue': 'Rating Value',
                                                 'geocode.latitude': 'Latitude',
                                                 'geocode.longitude': 'Longitude'})

reduced_geo_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': {'LocalAthorityName': '$LocalAuthorityName'},
                          'count': {'$sum': 1}}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'count': -1}}

# Form the pipepline
pipeline = [match_query, group_query, sort_values]

# Aggregate pipeline, cast as a list, and save in a variable
pipe_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of documents in pipe_results: {len(pipe_results)}")

# Print the first 10 results
pprint(pipe_results[:10])
```

```python
# Convert the result to a Pandas DataFrame
pipe_df = pd.json_normalize(pipe_results)

# Rename columns to match example in instructions
pipe_df = pipe_df.rename(columns={'_id.LocalAthorityName' : '_id'})

# Reorder columns to match instructions example
pipe_df = pipe_df[['_id', 'count']]

# Display the number of rows in the DataFrame
print(f"Number of rows in pipe_df: {len(pipe_df)}")
# Display the first 10 rows of the DataFrame
pipe_df.head(10)
```
